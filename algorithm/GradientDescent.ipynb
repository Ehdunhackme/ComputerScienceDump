```python
import numpy as np
import matplotlib.pyplot as plt

# Function to minimize
def function(x):
    return x**2

# Derivative of the function
def gradient(x):
    return 2 * x

# Gradient descent function
def gradient_descent(initial_x, learning_rate, num_iterations):
    x = initial_x  # start from an initial guess
    history = [x]  # record the path of x
    
    for i in range(num_iterations):
        grad = gradient(x)  # compute the gradient
        x = x - learning_rate * grad  # update the parameter
        history.append(x)  # keep track of the updates
    
    return history

# Parameters for gradient descent
initial_x = 10.0  # initial guess
learning_rate = 0.1  # learning rate
num_iterations = 20  # number of iterations

# Perform gradient descent
path = gradient_descent(initial_x, learning_rate, num_iterations)

# Visualization
x_values = np.linspace(-10, 10, 100)
y_values = function(x_values)

plt.plot(x_values, y_values, label='Function $f(x) = x^2$')
plt.plot(path, [function(x) for x in path], 'o-', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Simulation')
plt.legend()
plt.show()
```
